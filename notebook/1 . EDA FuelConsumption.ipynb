### 2) Data Collection
- Dataset Source - 
- YEAR: Integer type, representing the year of the vehicle.
- MAKE: Object type, representing the make or manufacturer of the vehicle.
- MODEL: Object type, representing the model of the vehicle.
- VEHICLE CLASS: Object type, representing the class of the vehicle.
- ENGINE SIZE: Float type, representing the engine size of the vehicle.
- CYLINDERS: Integer type, representing the number of cylinders in the engine.
- TRANSMISSION: Object type, representing the type of transmission.
- FUEL: Object type, representing the type of fuel used.
- FUEL CONSUMPTION: Float type, representing fuel consumption in some unit (probably liters per 100 km).
- HWY (L/100 km): Fuel consumption rate on highways, lower values indicate better efficiency.
- COMB (L/100 km): Combined fuel consumption rate (city and highway), lower values indicate better efficiency.
- COMB (mpg): Combined fuel efficiency in miles per gallon, higher values indicate better efficiency
- EMISSIONS: Integer type, representing emissions.

#### In TRANSMISSION 

- Manual: Manual transmissions (M1, M5, M6, M7).
- Basic Automatic: Basic automatic transmissions (A4, A5).
- Standard Automatic: Standard automatic transmissions (A6, A7, A8).
- Advanced Automatic: Advanced automatic transmissions (A9, A10, AS).
- Automated Manual: Automated manual transmissions (AM6, AM7, AM8).
- CVT: Continuously variable transmissions (AV1, AV6, AV7, AV8, AV10).

```python


# You can also create interaction features between existing numerical variables
df['ENGINE_SIZE_CYLINDERS'] = df['ENGINE SIZE'] * df['CYLINDERS']

# You can create binary features based on certain conditions
df['LARGE_ENGINE'] = df['ENGINE SIZE'].apply(lambda x: 1 if x > 3.0 else 0)

# Similarly, you can create categorical features based on ranges or conditions
def categorize_fuel_consumption(consumption):
    if consumption < 8:
        return 'Low'
    elif 8 <= consumption < 12:
        return 'Medium'
    else:
        return 'High'

df['FUEL_CONSUMPTION_CATEGORY'] = df['FUEL CONSUMPTION'].apply(categorize_fuel_consumption)

```

```python
import scipy.stats as stats

# Example of conducting statistical tests
# Let's assume we want to compare the mean emissions across different vehicle classes using ANOVA
anova_result = stats.f_oneway(df[df['VEHICLE CLASS'] == 'Compact']['EMISSIONS'],
                              df[df['VEHICLE CLASS'] == 'Mid-Size']['EMISSIONS'],
                              df[df['VEHICLE CLASS'] == 'SUV']['EMISSIONS'])

print("ANOVA results:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

```python
contingency_table = pd.crosstab(df['VEHICLE CLASS'], df['FUEL'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-square test results:")
print("Chi-square statistic:", chi2)
print("p-value:", p)
```

```python

# Let's assume we want to compare the mean emissions of vehicles with 4 cylinders vs. vehicles with 6 cylinders using a t-test
t_test_result = stats.ttest_ind(df[df['CYLINDERS'] == 4]['EMISSIONS'], df[df['CYLINDERS'] == 6]['EMISSIONS'])

print("\nIndependent t-test results:")
print("t-statistic:", t_test_result.statistic)
print("p-value:", t_test_result.pvalue)
```

```python
# Define a function to map existing categories to broader categories
def categorize_vehicle_class(class_name):
    if 'COMPACT' in class_name or 'SUBCOMPACT' in class_name or 'MINICOMPACT' in class_name or 'STATION WAGON - SMALL' in class_name or 'Subcompact' in class_name or 'Compact' in class_name or 'Station wagon: Small' in class_name:
        return 'small_car'
    elif 'MID-SIZE' in class_name or 'STATION WAGON - MID-SIZE' in class_name or 'Mid-size' in class_name:
        return 'medium_car'
    elif 'FULL-SIZE' in class_name or 'Full-size' in class_name:
        return 'large_car'
    elif 'PICKUP TRUCK' in class_name or 'Pickup truck: Standard' in class_name or 'Pickup truck: Small' in class_name:
        return 'pick_up_truck'
    else:
        return 'other'

# Apply the function to create a new column for broader categories
df['BROAD_VEHICLE_CLASS'] = df['VEHICLE CLASS'].apply(categorize_vehicle_class)

# Display the updated DataFrame
print(df[['VEHICLE CLASS', 'BROAD_VEHICLE_CLASS']].head())



print("----------------------------------------------------------------")



# Define a dictionary mapping abbreviations to full names
fuel_mapping = {
    'Z': 'Premium_gasoline',
    'D': 'Diesel',
    'X': 'Regular_gasoline',
    'E': 'Ethanol'
}

# Replace the abbreviations with their full names
df['FUEL'] = df['FUEL'].map(fuel_mapping)

# Display the updated DataFrame
print(df['FUEL'].unique())

```

```python
# Define a function to categorize transmission types
def categorize_transmission(transmission):
    if transmission.startswith('M'):
        return 'Manual'
    elif transmission.startswith('A4') or transmission.startswith('A5'):
        return 'Basic Automatic'
    elif transmission.startswith('A6') or transmission.startswith('A7') or transmission.startswith('A8'):
        return 'Standard Automatic'
    elif transmission.startswith('A9') or transmission.startswith('A10') or transmission.startswith('AS'):
        return 'Advanced Automatic'
    elif transmission.startswith('AM'):
        return 'Automated Manual'
    elif transmission.startswith('AV'):
        return 'CVT'
    else:
        return 'Other'

# Apply the function to create a new column for transmission groups
df['TRANSMISSION_GROUP'] = df['TRANSMISSION'].apply(categorize_transmission)

# Display the updated DataFrame with the transmission groups
print(df[['TRANSMISSION', 'TRANSMISSION_GROUP']].drop_duplicates())
```
